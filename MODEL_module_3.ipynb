```python
df.columns = ['id', 'city', 'cuisine', 'rank', 'R', 'price', 'reviews_count', 'reviews', 'url', 'ID']
#переименовываем колонки для удобства
```

```python
def change_cuisine1(i): #функция приведения кухонь к нижнему регистру, созданию списка через деление по запятой
    return i.lower().split(',')
df.cuisine = df.cuisine.dropna().apply(change_cuisine1)
```

```python
def change_cuisine2(i): #функция приведения списков кухонь к названиям без спец символов
    new_list = []
    for name in i:
        new_name = ''
        for letter in name:
            if letter in 'qwertyuiopasdfghjklzxcvbnm':
                new_name = new_name+letter
        new_list.append(new_name)
    return new_list
df.cuisine = df.cuisine.dropna().apply(change_cuisine2)
```

```python
unique_cuisine = [] #выводим в список все виды кухонь для последующего использования и общей оценки
for i in df.cuisine.dropna():
    for ii in i:
        if ii not in unique_cuisine:
            unique_cuisine.append(ii)
print('unique_cuisine:', unique_cuisine)
print()
print('len unique_cuisine:', len(unique_cuisine))
```

```python
frequent_cuisine = {} #определяем частоту встречаемости кухонь
for i in df.cuisine.dropna():
    for ii in i:
        if ii in frequent_cuisine:
            frequent_cuisine[ii] += 1
        else:
            frequent_cuisine[ii] = 1
print('frequent_cuisine:', frequent_cuisine)
```

```python
df.cuisine = df.cuisine.fillna('someone') #заполняем пропуски строковой величиной (предполагая одну кухню в пропусках)
```

```python
for i in unique_cuisine: #сделаем дамми с видами кухонь
    df[i] = 0
def find_cuisine(c):
    if i in c:
        return 1
    return 0
for i in unique_cuisine:
    df[i] = df.cuisine.apply(find_cuisine)
```

```python
def count_cuisine(i): #создание столбца количества кухонь в ресторане
    if i != 1:
        return len(i)
    return 1
df['count_cuisine'] = df.cuisine.apply(count_cuisine)
```

```python
def find_date(d): #находим все даты публикаций
    i = re.findall(r'\d\d.\d\d.\d{4}', d)
    return pd.to_datetime(i)
df['reviews_date'] = df.reviews.apply(find_date)
```

```python
df['reviews_latest'] = df['reviews_date'].apply(lambda x: x.max()) #находим самую позднюю публикацию
```

```python
df['reviews_latest'] = df['reviews_latest'].fillna(df['reviews_latest'].mean()) 
#среднее по позднему времени публикации для тех что без даты
```

```python
def count_of_days(d): #находим сколько прошло дней с последней публикации до сегодняшнего дня (проверим и такой признак)
    i = pd.to_datetime(datetime.date.today()) - d
    ii = str(i)
    iii = re.search(r'\d+', ii)
    return int(iii.group(0))
df['days_to_now'] = df['reviews_latest'].apply(count_of_days)
```

```python
df['tourists_vs_population'] = df.tourists_per_year / df.population #оценим соотношение количества туристов к населению
```

```python
df['reviews_count'] = df['reviews_count'].fillna(df['reviews_count'].mean())
#за пропуски в количестве отзывов возьмем среднее
```

```python
def change_price(s): #приведем оценку расценок к числовому признаку, пропуски как среднее
    if s == '$$ - $$$':
        return 2.5
    if s == '$$$$':
        return 4
    if s == '$':
        return 1
    return s 
df['price'] = df['price'].apply(change_price)
df['price'] = df['price'].fillna(df['price'].mean())
```

```python
df1 = df.groupby(['city'])['price'].mean() #найдем среднюю стоимость по городам
pd.DataFrame(df1)
```

```python
df = df.merge(df1, on='city') #объединяем с основным ДатаФреймом
```

# Разбиваем датафрейм на части, необходимые для обучения и тестирования модели

```python
# Х - данные с информацией о ресторанах, у - целевая переменная (рейтинги ресторанов)
X = df.drop(['id', 'R', 'city', 'cuisine', 'reviews', 'url', 'ID', 'reviews_date', 'reviews_latest'], axis = 1)
y = df['R']
X #выведем признаки по которым собираемся обучать, проверять модель
```

```python
# Загружаем специальный инструмент для разбивки:
from sklearn.model_selection import train_test_split
```

```python
# Наборы данных с меткой "train" будут использоваться для обучения модели, "test" - для тестирования.
# Для тестирования мы будем использовать 25% от исходного датасета.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

# Создаём, обучаем и тестируем модель

```python
# Импортируем необходимые библиотеки:
from sklearn.ensemble import RandomForestRegressor # инструмент для создания и обучения модели
from sklearn import metrics # инструменты для оценки точности модели
```

```python
# Создаём модель
regr = RandomForestRegressor(n_estimators=100)

# Обучаем модель на тестовом наборе данных
regr.fit(X_train, y_train)

# Используем обученную модель для предсказания рейтинга ресторанов в тестовой выборке.
# Предсказанные значения записываем в переменную y_pred
y_pred = regr.predict(X_test)
```

```python
# Сравниваем предсказанные значения (y_pred) с реальными (y_test), и смотрим насколько они в среднем отличаются
# Метрика называется Mean Absolute Error (MAE) и показывает среднее отклонение предсказанных значений от фактических.
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
```
